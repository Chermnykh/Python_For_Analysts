### 1.1. Скачать данные (недвижимость)

```python
df = pd.read_csv("kc_house_data.csv")
df.head()
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['price'], bins=20)
plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество домов');
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6, 4))
plt.hist(df['sqft_living'])
plt.title('Распределние квадратуры жилой площади')
plt.xlabel('Квадратура')
plt.ylabel('Количество домов');
```

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(6, 4))
plt.hist(df['yr_built'], bins=20)
plt.title('Распределние года постройки')
plt.xlabel('Год')
plt.ylabel('Количество домов');
```

## 2 задача

### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график  
Сделайте выводы

```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['no', 'yes']);
```

### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%', labels=data.index)
plt.title('Количество этажей');
```

### 2.2 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

### 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.  
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

На графике видна прямая зависимость между ценой и жилой площадью: чем больше площадь, тем выше цена

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Чем выше состояние дома, тем выше его стоимость

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Оценка')
plt.title('Распределение цены в зависимости от оценки');
```

Наибольшую стоимость имеют дома с оценкой дома = 13

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Этаж')
plt.title('Распределение цены в зависимости от количества этажей');
```

Наибольшую стоимость имеют дома с количеством этажей = 2.5

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Оценка вида')
plt.title('Распределение цены от оценки вида');
```

Наибольшую стоимость имеют дома с оценкой вида = 4
